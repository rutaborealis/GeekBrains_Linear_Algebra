### Урок 5. Сингулярное разложение матриц

```python
import math
import numpy as np
np.set_printoptions(precision=2, suppress=True)
```

#### Задание 1
Найти с помощью NumPy SVD для матрицы: <br>
$\begin{pmatrix}
1 & 2 & 0\\ 
0 & 0 & 5\\ 
3 & -4 & 2\\ 
1 & 6 & 5\\ 
0 & 1 & 0
\end{pmatrix}.$

Решение: <br>

```python
A = np.array([[1, 2, 0],
              [0, 0, 5],
              [3, -4, 2],
              [1, 6, 5],
              [0, 1, 0]])

print(f'Матрица A: \n{A}\n')

U, s, W = np.linalg.svd(A)

# Транспонируем матрицу W
V = W.T

# s - список диагональных элементов, приводим его к виду диагональной матрицы.
D = np.zeros_like(A, dtype=float)
D[np.diag_indices(min(A.shape))] = s

print(f'Матрица D:\n{D}\n')
print(f'Матрица U:\n{U}\n')
print(f'Матрица V:\n{V}\n')

# Проверка
print(f'Проверка (должны получить исходную матрицу А):\n{np.dot(np.dot(U, D), V.T)}')
```
