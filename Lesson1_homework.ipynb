#### Задание 1
Исследовать на линейную зависимость:
$$f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$$

Решение: <br>
Мы можем выразить одну функцию через другую, в данном случае:
$$f_{4}(x)=f_{3}(x)-f_{2}(x)-f_{1}(x)$$
следовательно, получаем линейную зависимость.

#### Задание 2
Исследовать на линейную зависимость:
$$f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2}.$$

Решение: <br>
Мы можем выразить одну функцию через другую, в данном случае:
$$f_{4}(x)=f_{3}(x)+2f_{2}(x)+0.5f_{1}(x)$$
следовательно, получаем линейную зависимость.

#### Задание 3
Найти координаты вектора $x = (2, 3, 5)\in \mathbb{R}^{3}$ в базисе $b_{1}=(0, 0, 10)$, $b_{2}=(2, 0, 0)$, $b_{3}=(0, 1, 0)$.

Решение: <br>
$x = (2, 3, 5) = (2, 0, 0) + (0, 3, 0) + (0, 0, 5) = 1 * (2, 0, 0) + 3 * (0, 1, 0) + 0.5 * (0, 0, 10) = 0.5b_{1} + 1*b_{2} + 3*b_{3} = (0.5, 1, 3).$

#### Задание 4
Найти координаты вектора $3x^{2}-2x+2\in\mathbb{R}^{3}[x]$: <br>
а) В базисе $1$, $x$, $x^{2}$; <br>
б) В базисе $x^{2}$, $x-1$, $1$.

Решение: <br>
а) Для базиса $1$, $x$, $x^{2}$ - координаты $(2, -2, 3)$. <br>
б) Для базиса $x^{2}$, $x-1$, $1$ приведем вектор к нужному виду: <br>
   $3x^{2}-2x+2 = 3x^{2} - 2(x-1)$. <br>
   Тогда для базиса $x^{2}$, $x-1$, $1$ - координаты $(3, -2, 0)$.

#### Задание 5
Установить, является ли линейным подпространством: <br>
а) Совокупность всех векторов трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю; <br>
б) Все векторы, являющиеся линейными комбинациями данных векторов $\{u_{1}, u_{2}, ..., u_{n}\}$.

Решение: <br>
а) Пример векторов: $a = (0, a2, a3)$, $b = (b1, 0, b3)$. <br>
   Сумма данных векторов: $a + b = (b1, a2, a3+b3)$. <br>
   Ни одна из двух первых координат не равна нулю, следовательно, данная совокупность векторов не является линейным
   подпространством. <br>
б) Пример векторов: $a = 3u_{1} + 5u_{2} + 7u_{3}$, $b = 2u_{4} + 6u_{5} + 8u_{6}$ <br>
   Сумма данных векторов $a + b = (3u_{1} + 5u_{2} + 7u_{3} + 2u_{4} + 6u_{5} + 8u_{6})$. <br>
   Умножение на скаляр вектора a: $K = 3$, $K * a = 9u_{1} + 15u_{2} + 21u_{3}$. <br>
   Следовательно, данная совокупность векторов является линейным подпространством.

```python
a = [4, 2, 4]
a_norm = np.linalg.norm(a)
a_norm
```

```python
b = [12, 3, 4]
b_norm = np.linalg.norm(b)
b_norm
```

#### Задание 3
Будет ли линейное пространство евклидовым, если за скалярное произведение принять: <br>
а) произведение длин векторов; <br>
б) утроенное обычное скалярное произведение векторов?

Решение: <br>
Линейное пространство над полем вещественных чисел называется _евклидовым пространством_, если на нем введено правило, согласно которому каждой упорядоченной паре векторов $x$ и $y$ поставлено в соответствие вещественное число, называемое _скалярным произведением_ и обозначаемое символом $(x, y)$. <br>
а) Не будет евклидовым пространством. <br>
б) Будет евклидовым пространством.

#### Задание 4
Какие из нижеперечисленных векторов образуют ортонормированный базис в линейном пространстве $\mathbb{R}^{3}$: <br>
а) $(1, 0, 0), (0, 0, 1);$ <br>
б) $(1/\sqrt{2}, -1/\sqrt{2}, 0), (1/\sqrt{2}, 1/\sqrt{2}, 0), (0, 0, 1);$ <br>
в) $(1/2, -1/2, 0), (0, 1/2, 1/2), (0, 0, 1);$ <br>
г) $(1, 0, 0), (0, 1, 0), (0, 0, 1)?$

```python
def orthonormal_basis(a, b, c):
    if np.dot(a, b) == np.dot(b, c) == np.dot(a, c) == (a[0] * b[0] * c[0]) + (a[1] * b[1] * c[1]) + (a[2] * b[2] * c[2]):
        return True
    else:
        return False
```

```python
a = [1 / np.sqrt(2), -1 / np.sqrt(2), 0]
b = [1 / np.sqrt(2), 1 / np.sqrt(2), 0]
c = [0, 0, 1]
orthonormal_basis(a, b, c)
```

```python
a = [1 / 2, -1 / 2, 0]
b = [0, 1 / 2, 1 / 2]
c = [0, 0, 1]
orthonormal_basis(a, b, c)
```

```python
a = [1, 0, 0]
b = [0, 1, 0]
c = [0, 0, 1]
orthonormal_basis(a, b, c)
```

Решение: <br>
а) Векторы не образуют ортонормированный базис, т. к. их всего 2. <br>
б) Да, образуют. <br>
в) Нет, не образуют. <br>
г) Да, образуют.
